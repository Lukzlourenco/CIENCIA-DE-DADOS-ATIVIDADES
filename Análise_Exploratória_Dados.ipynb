```python
#Filtros
# Busca apenas aqueles que possuem 17 anos
df[df.Idade == 17]
```

```python
# Busca pessoas com mais de 17 anos 
df[df.Idade > 17]
```

```python
#Busca pessoas de todas as idades, menos as com 18 anos 
df[df.Idade != 18]
```

```python
#Pessoas com mais de 17 e menos que 20 anos
df[(df.Idade > 17) & (df.Idade<20)]
```

```python
#Classificação 
df.sort_values(by = ['Sexo'])
```

```python
df.sort_values(by = ['Sexo' , 'Idade'])
```

```python
df.sort_values(by = ['Idade'] , ascending = False)
```

```python
#Histograma com a altura das das pessoas do dataframe
plt.hist(df['Alt'], bins=8)
plt.show() # O "plt.show()" oculta os metadados 
```

```python
#Histograma com a idade das pessoas analisadas
plt.hist(df['Idade'], bins=8)
plt.show()
```

```python
# Usando um grafico de pizza por sexo e idade
df.groupby(['Sexo'])['Idade'].count().plot(kind='pie')
plt.show()
```

```python
df.groupby(['Sexo'])['Idade'].count().plot(kind='barh')
plt.show()
```

```python
#O uso de seaborn para mostrar o Gráfico de Brarras
sns.set_style("whitegrid")
ax = sns.barplot(x='Sexo', y='Idade', data=df, estimator=len)
```
